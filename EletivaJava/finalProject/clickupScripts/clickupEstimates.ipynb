```python

total_horas = 0
for task in data.get("tasks", []):
    time_estimate = task.get("time_estimate")
    if time_estimate:
        horas = time_estimate / (1000 * 60 * 60)  # ms → h
        total_horas += horas
        print(f"{task['name']}: {horas:.2f}h")

print(f"\nTotal estimado: {total_horas:.2f} horas")
print(f"Custo total: R$ {total_horas * VALOR_HORA:.2f}")
```

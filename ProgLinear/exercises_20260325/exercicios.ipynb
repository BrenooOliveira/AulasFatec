# Problema 1 — Modelagem com Programação Linear

Um sapateiro faz **6 sapatos/hora** (se dedicar 100% do tempo a sapatos) e **5 cintos/hora** (se dedicar 100% do tempo a cintos).
Cada sapato consome 2 unidades de couro e cada cinto consome 1 unidade. Há no máximo 6 unidades de couro por hora.
Lucro unitário: sapato = 5, cinto = 2.

## Variáveis de decisão
- $x_1$: fração da hora dedicada à produção de sapatos
- $x_2$: fração da hora dedicada à produção de cintos

Como são frações de tempo: $x_1, x_2 \ge 0$.

## Função objetivo (maximizar lucro/hora)
Lucro por hora se dedicar 100% a sapatos: $6 \cdot 5 = 30$.
Lucro por hora se dedicar 100% a cintos: $5 \cdot 2 = 10$.

$$\max Z = 30x_1 + 10x_2$$

## Restrições
1. Tempo da hora disponível:
$$x_1 + x_2 \le 1$$

2. Couro disponível por hora:
- Sapatos: $6x_1$ pares/hora, cada par usa 2 de couro $\Rightarrow 12x_1$
- Cintos: $5x_2$ unidades/hora, cada unidade usa 1 de couro $\Rightarrow 5x_2$

$$12x_1 + 5x_2 \le 6$$

3. Não negatividade:
$$x_1, x_2 \ge 0$$


```python
from scipy.optimize import linprog

# Função Objetivo: 30x1 + 10x2 (lucro máximo)
fo = [-30, -10] # negativo = maximizar
```

```python
# Restrições: Cada uma vira uma linha da matriz `A_ub`:
# Retrição de hora: x1 + x2 <= 1
# Restrição de Couro:  Em 1 hora dedicada inteiramente a sapatos ele faz 6 sapatos, cada um consumindo 2 unidades de couro = 12 unidades. Para cintos: 5 × 1 = 5 unidades: 12·x1 + 5·x2 ≤ 6

A_ub = [
    [1, 1], # coef de restricoes de tempo
    [12, 5] # coef de restricoes de couro
]

b_ub = [1, 6] # limites das restricoes

# A_ub @ x <= b_ub, ou seja, cada linha de A multiplicada pelo vetor x deve ser ≤ ao valor correspondente em b.
```

```python
# x1 ≥ 0 e x2 ≥ 0 (não faz sentido tempo negativo)
bounds = [(0, None), (0, None)]
```

```python
result = linprog(fo, A_ub=A_ub, b_ub = b_ub,bounds=bounds, method='highs')
```

```python
x1, x2 = result.x
print(f"Tempo em sapatos: {x1*100:.1f}%  → {6*x1:.2f} sapatos/hora")
print(f"Tempo em cintos:  {x2*100:.1f}%  → {5*x2:.2f} cintos/hora")
print(f"Lucro máximo: R$ {-result.fun:.2f}/hora")
```

# Problema 2 — Planejamento mensal de produção (P1 e P2)

Uma empresa produz dois itens: **P1** e **P2**.
- Lucro unitário: P1 = 100 u.m.; P2 = 150 u.m.
- Tempo por unidade: P1 = 2 h; P2 = 3 h
- Tempo mensal disponível: 120 h
- Limites de demanda: P1 ≤ 40 e P2 ≤ 30

## Variáveis de decisão
- $x_1$: quantidade mensal produzida de P1 (unidades)
- $x_2$: quantidade mensal produzida de P2 (unidades)

## Função objetivo
Maximizar o lucro total mensal:

$$\max Z = 100x_1 + 150x_2$$

## Restrições
1. Tempo disponível:

$$2x_1 + 3x_2 \le 120$$

2. Limite de demanda de P1:

$$x_1 \le 40$$

3. Limite de demanda de P2:

$$x_2 \le 30$$

4. Não negatividade:

$$x_1, x_2 \ge 0$$



```python


## Função objetivo (linprog minimiza):
c = [-100, -150]

## Restrições no formato A_ub @ x <= b_ub
## 2x1 + 3x2 <= 120   (tempo)
## x1 <= 40           (demanda de P1)
## x2 <= 30           (demanda de P2)
A_ub = [
    [2, 3],
    [1, 0],
    [0, 1]
 ]
b_ub = [120, 40, 30]

## Não negatividade
bounds = [(0, None), (0, None)]

## Resolver
resultado_p2 = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method="highs")

## Exibir solução
x1_p2, x2_p2 = resultado_p2.x
lucro_p2 = -resultado_p2.fun

print(f"Produção ótima de P1: {x1_p2:.2f} unidades/mês")
print(f"Produção ótima de P2: {x2_p2:.2f} unidades/mês")
print(f"Lucro máximo mensal: {lucro_p2:.2f} u.m.")
print(f"Status do solver: {resultado_p2.message}")
```

# Problema 3 — Carregamento do caminhão de frutas

Um vendedor pode transportar até **800 caixas** para sua região de vendas.
Condições:
- Laranjas: transportar **exatamente 200 caixas**, lucro de 20 u.m./caixa
- Pêssegos: transportar **pelo menos 100 caixas**, lucro de 10 u.m./caixa
- Tangerinas: transportar **no máximo 200 caixas**, lucro de 30 u.m./caixa

## Variáveis de decisão
- $x_1$: caixas de laranjas
- $x_2$: caixas de pêssegos
- $x_3$: caixas de tangerinas

## Função objetivo
Maximizar lucro total:

$$\max Z = 20x_1 + 10x_2 + 30x_3$$

## Restrições
1. Capacidade do caminhão:

$$x_1 + x_2 + x_3 \le 800$$

2. Laranjas (fixo):

$$x_1 = 200$$

3. Pêssegos (mínimo):

$$x_2 \ge 100$$

4. Tangerinas (máximo):

$$x_3 \le 200$$

5. Não negatividade:

$$x_1, x_2, x_3 \ge 0$$

```python

# Problema 3
## x = [x1, x2, x3] = [laranjas, pessegos, tangerinas]

# Função objetivo (max -> min)
c = [-20, -10, -30]

# Capacidade total do caminhão
A_ub = [[1, 1, 1]]
b_ub = [800]

# Limites individuais
# x1 = 200 (exatamente)
# x2 >= 100
# 0 <= x3 <= 200
bounds = [(200, 200), (100, None), (0, 200)]

resultado_p3 = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method="highs")

x1_p3, x2_p3, x3_p3 = resultado_p3.x
lucro_p3 = -resultado_p3.fun

print(f"Laranjas:   {x1_p3:.0f} caixas")
print(f"Pêssegos:   {x2_p3:.0f} caixas")
print(f"Tangerinas: {x3_p3:.0f} caixas")
print(f"Lucro máximo: {lucro_p3:.2f} u.m.")
print(f"Status do solver: {resultado_p3.message}")
```

# Problema 4 — Programação semanal da TV

A rede tem dois programas:
- Programa A: 20 min de música + 1 min de propaganda, 30.000 telespectadores por exibição
- Programa B: 10 min de música + 1 min de propaganda, 10.000 telespectadores por exibição

Condições semanais:
- Pelo menos 5 minutos de propaganda no total
- No máximo 80 minutos de música

## Variáveis de decisão
- $x_1$: número de exibições semanais do programa A
- $x_2$: número de exibições semanais do programa B

## Função objetivo
Maximizar audiência semanal:

$$\max Z = 30000x_1 + 10000x_2$$

## Restrições
1. Música disponível:

$$20x_1 + 10x_2 \le 80$$

2. Propaganda mínima (1 min por exibição):

$$x_1 + x_2 \ge 5$$

3. Não negatividade:

$$x_1, x_2 \ge 0$$


```python

# Problema 4
## x = [x1, x2] = [exibicoes de A, exibicoes de B]

# Função objetivo (max -> min)
c = [-30000, -10000]

# Restrições no formato A_ub @ x <= b_ub
# 20x1 + 10x2 <= 80 (música)
# x1 + x2 >= 5  ->  -x1 - x2 <= -5 (propaganda mínima)
A_ub = [
    [20, 10],
    [-1, -1]
 ]
b_ub = [80, -5]

bounds = [(0, None), (0, None)]

resultado_p4 = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method="highs")

x1_p4, x2_p4 = resultado_p4.x
audiencia_p4 = -resultado_p4.fun

print(f"Programa A por semana: {x1_p4:.2f}")
print(f"Programa B por semana: {x2_p4:.2f}")
print(f"Audiência máxima semanal: {audiencia_p4:.0f} telespectadores")
print(f"Status do solver: {resultado_p4.message}")
```

# Problema 5 — Produção diária de cintos M1 e M2

Dados do problema:
- M1 usa o dobro do tempo de M2
- Se só produzisse M2: capacidade de 1000 unidades/dia (limite de tempo)
- Couro disponível: até 800 cintos/dia (M1 + M2)
- Fivelas: no máximo 400 para M1 e 700 para M2 por dia
- Lucros unitários: M1 = 4 u.m.; M2 = 3 u.m.

## Variáveis de decisão
- $x_1$: quantidade diária de cintos M1
- $x_2$: quantidade diária de cintos M2

## Função objetivo
Maximizar lucro diário:

$$\max Z = 4x_1 + 3x_2$$

## Restrições
1. Tempo disponível (M1 consome o dobro do tempo de M2):

$$2x_1 + x_2 \le 1000$$

2. Couro disponível:

$$x_1 + x_2 \le 800$$

3. Fivelas:

$$x_1 \le 400$$

$$x_2 \le 700$$

4. Não negatividade:

$$x_1, x_2 \ge 0$$



```python
# Problema 5
## x = [x1, x2] = [M1, M2]

# Função objetivo (max -> min)
c = [-4, -3]

# Restrições A_ub @ x <= b_ub
A_ub = [
    [2, 1],   # tempo
    [1, 1],   # couro
    [1, 0],   # fivelas M1
    [0, 1]    # fivelas M2
 ]
b_ub = [1000, 800, 400, 700]

bounds = [(0, None), (0, None)]

resultado_p5 = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method="highs")

x1_p5, x2_p5 = resultado_p5.x
lucro_p5 = -resultado_p5.fun

print(f"Produção ótima de M1: {x1_p5:.2f} unidades/dia")
print(f"Produção ótima de M2: {x2_p5:.2f} unidades/dia")
print(f"Lucro máximo diário: {lucro_p5:.2f} u.m.")
print(f"Status do solver: {resultado_p5.message}")
```
